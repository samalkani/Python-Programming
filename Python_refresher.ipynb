```python
# write a function that creates a random list of 10 integers
# between 0 and 100
import random
random_list = []
def create_list():
    for i in range(10):
      # this is how you add an element at the end of a list
      random_list.append(random.randint(0,100))
      

create_list()
random_list
```

```python
# write a function that class those numbers from smallest to biggest
def order_list(random_list):

      # we need to initiate the list as empty
      ordered_list = []

      # for each element, we are going to check if it is smaller or bigger
      # than the precedent element
      for number in random_list:

        # if we are dealing with the first number, we automatically
        # add it to the list.

        if len(ordered_list) == 0:
            ordered_list.append(number)

        else:
          # Otherwise we compare the last number of our
          # ordered list and if the new number is bigger, then we append it.
          # We access the last number of our list by getting the length - 1.

          if number > ordered_list[len(ordered_list)-1]:
              ordered_list.append(number)
          else:
            # we need to find where to put the number in our ordered list
            # so we compare it with others

            for compared_number in ordered_list:
                if number < compared_number:
                    index = ordered_list.index(compared_number)
                    # we insert the new number at its position

                    ordered_list.insert(index, number)
                    # we stop the for loops
                    break

      return ordered_list

order_list(random_list)
```
